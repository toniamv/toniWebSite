```python
import matplotlib.pyplot as plt

# Dados
anos = [2021, 2022, 2023, 2024, 2025]

notas = {
    "Ciência da Computação": {
        "valores": [736.94, 742.94, 745.75, 765.48, 773.50],
        "cor": "#1f77b4"  # azul
    },
    "Engenharia Aeronáutica": {
        "valores": [715.82, 693.72, 725.79, 734.81, 740.13],
        "cor": "#ff7f0e"  # laranja
    },
    "Medicina": {
        "valores": [654.52, 678.37, 686.59, 708.42, 699.96],
        "cor": "#2ca02c"  # verde
    }
}

# Estilo limpo
plt.style.use('ggplot')  # Estilo limpo, já incluso no matplotlib

# Criar figura
plt.figure(figsize=(10, 6))

# Plotar cada linha
for curso, dados in notas.items():
    plt.plot(anos, dados["valores"], marker='o', label=curso, color=dados["cor"], linewidth=2)

# Títulos e legendas
plt.title('Nota de Corte SISU (Ampla Concorrência) por Ano', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Nota (Ano)', fontsize=12)
plt.xticks(anos)  # Mostrar anos como inteiros
plt.legend()
plt.tight_layout()

# Exibir
plt.show()
```

```python
import matplotlib.pyplot as plt

# Anos
anos = [2021, 2022, 2023, 2024, 2025]

# Dados das notas
notas = {
    "Ciência da Computação": {
        "valores": [736.94, 742.94, 745.75, 765.48, 773.50],
        "cor": "#1f77b4"  # azul
    },
    "Engenharia Aeronáutica": {
        "valores": [736.92, 720.40, 661.47, 765.48, 800.88],
        "cor": "#ff7f0e"  # laranja
    },
    "Medicina": {
        "valores": [795.38, 790.75, 786.23, 799.55, 799.27],
        "cor": "#2ca02c"  # verde
    }
}

# Estilo visual
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))

# Plotar linhas
for curso, dados in notas.items():
    plt.plot(anos, dados["valores"], marker='o', label=curso, color=dados["cor"], linewidth=2)

# Títulos e eixos
plt.title('Nota de Corte SISU (Ampla Concorrência) por Ano', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Nota(ano) (Nota de Corte Ampla do Sisu no Ano)', fontsize=12)
plt.xticks(anos)  # Mostrar anos como inteiros
plt.legend()
plt.tight_layout()

# Exibir gráfico
plt.show()
```
